##Plot for HW8
####CO2 emissions per capita against population density for US counties with over 100 thousand people
Data obtained from Vulcan Project Data and US Census

```python
#reading in previously cleande and prepared data
mymsa = pd.read_csv('/Users/mariaortiz/PUI2015/CAPS2/dataclean.csv')
#mymsa.head(5)

#pop total mean over co2 per capita
result = mymsa.groupby('CBSA Code').agg({'popest2002':mean,'co2total':sum})
result.shape
#result.head()

#running regression
import statsmodels.formula.api as smf
from statsmodels.graphics.api import abline_plot 
# Create a dictionary so we can use formula.ols
d = {'x': result['popest2002'], 'y': result['co2total']}

# Run regression and print summary
mod = smf.ols(formula='y ~ x', data = d).fit()
#mod.summary()

#Plotting
figure, ax = pl.subplots(figsize=(7,7))
x = result['popest2002']
y = result['co2total']
ax.loglog(x, y, '.')

# Plot regression line
abline_plot(model_results=mod, color='green', ax=ax, lw = 3)

ax.set_title('Linear regression showing upwards trend')
pl.ylabel('CO2 Total')
pl.xlabel('Population 2002')
ax.set_xlim=(0,2000000)
ax.set_ylim(0,60)
ax.legend(loc=2)
```

```python
#histogram and line fit, random data with poisson
import scipy
data = np.random.poisson(5, 10000)
mean, sd = scipy.stats.norm.fit(data)

#generating pdf to plot it
x = np.linspace(-5,25,100)
y = scipy.stats.norm.pdf(x, mean, sd)

figure, ax = pl.subplots(figsize=(10,10))

ax.hist(data, normed=True, bins=10, color='darkslateblue')

ax.plot(x, y, 'g--')
```
