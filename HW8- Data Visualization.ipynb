```python
#histogram and line fit, random data with poisson
import scipy
data = np.random.poisson(5, 10000)
mean, sd = scipy.stats.norm.fit(data)

#generating pdf to plot it
x = np.linspace(-5,25,100)
y = scipy.stats.norm.pdf(x, mean, sd)

figure, ax = pl.subplots(figsize=(10,10))

ax.hist(data, normed=True, bins=10, color='darkslateblue')

ax.plot(x, y, 'g--')
```
